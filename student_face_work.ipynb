```python
train_data=[]
target=[]
path_image= r"C:\Users\Student\Desktop\dataset_1\students_face"
with open(r'C:\Users\Student\Desktop\dataset_1\students_face\train.csv','rt')as dat:
    data = csv.reader(dat, delimiter = ',')
    for row in data:
        #print(row)
        temp=cv2.imread(os.path.join(path_image, 'train', row[0]))
        img = cv2.resize(temp,(50,50))         
        train_data.append(img)
        target.append(row[1])
       
```

```python
train_data=np.array(train_data)
train_data1=train_data/255.
target=np.array(target)
uni=np.unique(target)
print(uni)
```

```python
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = target.reshape(len(target), 1)
Y = onehot_encoder.fit_transform(integer_encoded)
print(Y)
```

```python
def IntelNet(input_shape = (50, 50, 3), classes = len(uni)):
    # Define the input as a tensor with shape input_shape
    X_input = Input(input_shape)
    
    # Zero-Padding
    X = ZeroPadding2D((3, 3))(X_input)
    
    
     # Stage 1
    X = Conv2D(128, (5, 5), strides = (4, 4), name = 'conv1',padding = 'SAME')(X)
    X = BatchNormalization(axis = 3, name = 'bn_conv1')(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((6, 6), strides=(4, 4),padding = 'SAME')(X)
    

    X = Conv2D(128, (7, 7), strides = (3, 3), name = 'conv2')(X)
    X = BatchNormalization(axis = 3, name = 'bn_conv2')(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((6, 6), strides=(4, 4))(X)
    
  #  X = Conv2D(128, (1, 11), strides = (1, 1), name = 'conv3')(X)
   # X = BatchNormalization(axis = 3, name = 'bn_conv3')(X)
    #X = Activation('relu')(X)
    
    #X = Conv2D(256, (1, 12), strides = (1, 1), name = 'conv4')(X)
    #X = BatchNormalization(axis = 3, name = 'bn_conv4')(X)
    #X = Activation('relu')(X)
   # X = MaxPooling2D((3, 3), strides=(2, 2))(X)
    
    
    
    
    
    
    
    
    # output layer
    X = Flatten()(X)
    X = Dense(classes, activation='softmax', name='fc' + str(classes), kernel_initializer = glorot_uniform(seed=0))(X)
    
    
    # Create model
    model = Model(inputs = X_input, outputs = X, name='ResNet50')

    return model
```

```python
model = IntelNet(input_shape = (50, 50, 3), classes = len(uni))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(train_data,Y,test_size = 0.2)
```

```python
model.fit(X_train, Y_train, epochs = 100, batch_size = 100)
```

```python
model.evaluate(x=X_test, y=Y_test, batch_size=10)
```
